```python
import pandas as pd
```

```python
# Creating a DataFrame with missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", None, "Eve"],
    "Age": [25, None, 30, 22, None],
    "City": ["New York", "Los Angeles", None, "Chicago", "Houston"],
    "Salary": [50000, 60000, None, None, 75000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
df.dropna()
```

```python
df.dropna(axis = 0)            
```

```python
df.dropna(axis = 1)            
```

```python
df.dropna(axis = 1, how = 'any')
```

```python
df.dropna(how = 'all')     
```

```python
df.dropna(thresh = 1)
```

```python
df.dropna(thresh = 4)
```

```python
df.dropna(subset = ['Salary'])    
```

```python
#fillna
```

```python
df.fillna(value = 2) 
```

```python
df.fillna(2)
```

```python
df.fillna({'Name': 'Zainab', 'Age' : '0'})  
```

```python
df.fillna(value={'Name': 'Zainab', 'Age' : '0'})  
```

```python
#To fill values by forward fill:
```

```python
df.fillna(method = 'ffill')    
```

```python
df.fillna(method = 'bfill')    
```

```python
df.fillna(9, limit = 3)  
```

```python
df.fillna(method = 'bfill', axis=0)    
```

```python
df.fillna(5, inplace = True)  
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.	Read the dataset
```

```python
data=pd.read_csv(r"C:\Users\miari\Downloads\gene_expression_file.csv")
data
```

```python
# 2.Explore the data:
```

```python
print(f"Number of rows (genes): {data.shape[0]}")
print(f"Number of columns (samples): {data.shape[1]}\n")
print("First few rows of the dataset:")
data.head(2)
```

```python
# 3.Clean the data:
```

```python
clean_data=data.dropna()
print("\nAfter cleaning, number of rows:", clean_data.shape[0])
clean_data
```

```python
# 4.Basic statistics:
```

```python
# Remove unnecessary columns
columns_to_remove = ['Gene Name'] 
final_data = clean_data.drop(columns=columns_to_remove)

# Calculate statistics for each row
mean_values = np.mean(final_data, axis=1)
median_values = np.median(final_data, axis=1)
mode_values = final_data.mode(axis=1)[0]  # Mode might return multiple columns; [0] selects the first mode

# Add calculated statistics back to the dataset
clean_data['Mean'] = mean_values
clean_data['Median'] = median_values
clean_data['Mode'] = mode_values

# Display statistics with gene names
print("\nMean, Median, and Mode values for each gene:")
print(clean_data[['Gene Name', 'Mean', 'Median', 'Mode']])
```

```python
# 5.Identify top genes:
```

```python
clean_data['Mean Expression'] = mean_values
top_genes = clean_data.nlargest(3, 'Mean')[['Gene Name', 'Mean']]
print("\nTop 3 highly expressed genes:")
print(top_genes)
```

```python
# 6.Analyse specific genes:
```

```python
# Analyze specific genes
gene_name = input("\nEnter a gene name for detailed analysis: ")
if gene_name in clean_data['Gene Name'].values:
    gene_data = clean_data[clean_data['Gene Name'] == gene_name].iloc[0, 1:-3] # "iloc"select rows and columns from a DataFrame by their integer-based positions
    print(f"\nExpression values for {gene_name}:")
    print(gene_data)

# Plot the expression values
    plt.figure(figsize=(12, 6))
    plt.plot(gene_data.index, gene_data.values, marker='s', label=gene_name, color="purple")
    plt.xticks(ticks=range(0, len(gene_data.index), 1), rotation=90, fontsize=8)  # sets the x-axis tick values,with a step of 5

    plt.title(f"Expression Levels of {gene_name}")
    plt.xlabel("Samples")
    plt.ylabel("Gene Expression")
    plt.grid(True)
    plt.legend()
    plt.show()
else:
    print(f"\nGene '{gene_name}' not found in the dataset.")
```
